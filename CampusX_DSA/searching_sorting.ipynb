Linear Search

```python
def linear_search(arr, item):
  for i in range(len(arr)):
    if arr[i] == item:
      return i
      i += 1
  return -1
```

```python
L = [1,6,2,3,7,8,90,12,34,43,18]
```

```python
linear_search(L, 90)
```

```python
linear_search(L, 9)
```

```python
def binary_search(arr, item):
  low = 0
  high = len(arr) - 1

  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == item:
      return mid
    elif item > arr[mid]:
      low = mid + 1
    else: # item < arr[mid]
      high = mid - 1
  return -1
```

```python
L = [10, 20, 30 , 50, 90, 100]
```

```python
binary_search(L, 90)
```

```python
binary_search(L, 10)
```

```python
binary_search(L, 100)
```

detecting if a given array is sorted or not

```python
def is_sorted(arr):
  sorted = True
  for i in range(len(arr) -  1):
    if arr[i] > arr[i+1]:
      sorted =  False
  return sorted
```

Bubble Sort

- in sorting time complexity is measured in the terms of comparisons
- n - 1 possible comparisons
- n-1 possible swaps
- O(n^2) TC

Bubble sort is called 'bubble sort' because of the way larger elements 'bubble up' to their correct positions at the end of the array with each pass. Imagine bubbles in a glass of water – the lighter ones rise to the top. Similarly, in bubble sort, the larger (or 'heavier') elements gradually move towards the end of the list, just like bubbles rising to the surface.



```python
def bubble_sort(arr):
  for i in range(len(arr) - 1):
    for j in range(len(arr) - 1 - i):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
```

```python
L = [1,6,2,3,7,8,90,12,34,43,18]
```

```python
bubble_sort(L)
```

## Stable and Adaptive Sorting Algorithms

### Stable Sorting Algorithms
A sorting algorithm is said to be **stable** if two objects with equal keys appear in the same order in the sorted output as they appeared in the input array. In other words, if `a` and `b` are two elements with the same key, and `a` appears before `b` in the original array, then `a` must appear before `b` in the sorted array.

**Examples of Stable Sorts:**
*   Bubble Sort
*   Insertion Sort
*   Merge Sort
*   Counting Sort

**Examples of Unstable Sorts:**
*   Selection Sort
*   Quick Sort
*   Heap Sort

### Adaptive Sorting Algorithms
A sorting algorithm is said to be **adaptive** if its time complexity improves when the input array is already partially sorted. The more sorted the input is, the faster the algorithm performs. Non-adaptive algorithms perform the same number of operations regardless of the input's sortedness.

**Examples of Adaptive Sorts:**
*   Insertion Sort (highly adaptive)
*   Bubble Sort (adaptive, can be optimized for already sorted arrays)
*   TimSort (used in Python and Java, highly adaptive hybrid sort)

**Examples of Non-Adaptive Sorts:**
*   Selection Sort
*   Heap Sort
*   Merge Sort (generally, though some variants can be adaptive)
*   Quick Sort (generally, though some variants can be adaptive)


**Why are these properties important?**
*   **Stability** is crucial when elements have associated data that needs to maintain its relative order (e.g., sorting a list of students by name, then by grade; if the name sort is stable, students with the same name retain their original order).
*   **Adaptivity** is beneficial when dealing with data that is often nearly sorted, as it can lead to significant performance improvements.

we can make our bubble sorting code adaptive by the following

```python
def bubble_sort(arr):
  for i in range(len(arr) - 1):
    flag = 0
    for j in range(len(arr) - 1 - i):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        flag = 1
    if flag == 0:
      break
  print(arr)
```

```python
def selection_sort(arr):
  for i in range(len(arr) - 1):
    min = i
    for j in range(i+1, len(arr)):
      if arr[j] < arr[min]:
        min = j
    arr[i], arr[min] = arr[min], arr[i]

  print(arr)
```

```python
L = [1,6,2,3,7,8,90,12,34,43,18]
```

```python
selection_sort(L)
```

space complexity is O(1) and time complexity is O(n^2)

Merge Sort

```python
def merge_sorted(arr1, arr2):
  i = j = 0
  merged = []
  while i < len(arr1) and j < len(arr2):
    if arr1[i] < arr2[j]:
      merged.append(arr1[i])
      i += 1
    else:
      merged.append(arr2[j])
      j += 1

  while i < len(arr1):
    merged.append(arr1[i])
    i += 1
  while j < len(arr2):
    merged.append(arr2[j])
    j += 1

  return merged
```

```python
def merge_sort(arr):
    if len(arr) == 1:
        return arr

    mid = len(arr)//2

    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge_sorted(left, right)
```

```python
def merge_sorted(arr1, arr2, arr):
  i = j = k = 0
  while i < len(arr1) and j < len(arr2):
    if arr1[i] < arr2[j]:
      arr[k] = arr1[i]
      i += 1
    else:
      arr[k] = arr2[j]
      j += 1
    k += 1

  while i < len(arr1):
    arr[k] = arr1[i]
    i += 1
    k += 1
  while j < len(arr2):
    arr[k] = arr2[j]
    j += 1
    k += 1

  return
```

```python
def merge_sort(arr):
    if len(arr) == 1:
        return arr

    mid = len(arr)//2

    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    merge_sorted(left, right, arr)
    return arr
```
