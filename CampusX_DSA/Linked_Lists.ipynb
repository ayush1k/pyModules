```python
class LinkedList:

  def __init__(self):
    # empty linked list
    self.head = None
    # n tells the number of nodes in the LL
    self.n = 0

  def __len__(self):
    return self.n

  def insert_head(self, value):
    # new node
    new_node = Node(value)
    # create connection
    new_node.next = self.head
    # reassign head
    self.head =  new_node
    # increment n
    self.n = self.n + 1

  def __str__(self):
    curr = self.head
    result = ''
    while curr != None:
      result = result + str(curr.data) + '->'
      curr = curr.next

    return result[:-2]

  def append(self, value):
    # new node
    new_node = Node(value)

    if self.head == None:
      # means the list is empty
      self.head = new_node
      self.n = self.n + 1
      return

    curr = self.head
    while curr.next != None:
      # agar item k exactly upar ja kr rukna hai tho ye loop chalega
      curr = curr.next
    curr.next = new_node
    self.n = self.n + 1


  def insert_after(self,after,value):
    new_node = Node(value)
    curr = self.head

    while curr != None:
      if curr.data == after:
        new_node.next = curr.next
        curr.next = new_node
        self.n = self.n + 1
        return
      curr = curr.next

    # case 1 - item found and you are coming when break statement is executed - item aapko mil gaya
    if curr != None:
      return 'Item not found'
    # case 2 - break nai hua - loop pura chala - item nahi mila - mtlb ki curr = None
    return 'Item not found'

  def clear(self):
    self.head = None
    self.n = 0

  def delete_head(self):
    if self.head == None:
      return 'Empty Linked List'

    self.head =  self.head.next
    self.n = self.n -1

  def pop(self):
    if self.head == None:
      return 'Empty Linked List'

    curr = self.head

    # kya linked list me ek item hai
    if curr.next == None:
      return self.delete_head()

    while curr.next.next != None:
      curr = curr.next
    # curr has the value of the second last item
    curr.next = None
    self.n = self.n - 1

  def remove(self, value):
    if self.head == None:
      return 'Empty Linked List'

    if self.head.data == value:
      return self.delete_head()

    curr = self.head
    while curr.next != None:
      if curr.next.data == value:
        break
      curr = curr.next
    # case 1 - item found
    # case 2 - item not found
    if curr.next == None:
      # item nahi mila
      return 'Item not found'
    else:
      # item mil gaya
      curr.next = curr.next.next
      self.n = self.n - 1

  def search(self, item):
    if self.head == None:
      return 'Empty Linked List'
    curr = self.head
    pos = 0
    while curr != None:
      if curr.data == item:
        return pos
      curr = curr.next
      pos = pos + 1

    return 'Not found'

  def __getitem__(self,index):
    curr = self.head
    pos = 0

    while curr != None:
      if pos == index:
        return curr.data
      curr = curr.next
      pos = pos + 1

    return 'Index Error'

  def find_max(self):
    max_value = -1
    curr = self.head
    while curr != None:
      if curr.data > max_value:
        max_value = curr.data
        curr = curr.next
      else:
        curr = curr.next

    return max_value

  def replace_max(self, value):
    temp = self.head
    max = temp
    while temp != None:
      if temp.data > max.data:
        max = temp
      temp = temp.next

    max.data = value

  def sum_odd_nodes(self):
    curr = self.head
    counter = 0
    sum = 0
    while curr != None:
      if counter %2 != 0:
        sum = sum + curr.data
      counter = counter + 1
      curr = curr.next

    return sum

  def reverse(self):
    # inplace reversal
    curr_node = self.head
    prev_node = None

    while curr_node != None:
      next_node = curr_node.next
      curr_node.next = prev_node
      prev_node = curr_node
      curr_node = next_node

    self.head = prev_node

  def create_new_sentence(self):
    curr = self.head
    result = ''
    while curr != None:
      if curr.data == '*' or curr.data == '/':
        curr.data = " "
        if curr.next.data == '*' or curr.next.data == '/':
          curr.next.next.data = curr.next.next.data.upper()
          curr.next = curr.next.next
      result = result + curr.data
      curr = curr.next
    return result
```
