# implementing stacks via Arrays

```python
class Stack:
  def __init__(self,size):
    self.size = size
    self.stack = [None] * self.size
    self.top = -1

  def push(self, value):
    if self.top == self.size - 1:
      return 'Overflow'
    else:
      self.top += 1
      self.stack[self.top] = value

  def pop(self):
    if self.top == -1:
      return 'Stack empty'
    else:
      popped_value = self.stack[self.top]
      self.stack[self.top] = None # Optional: clear the value at the old top position
      self.top -= 1
      return popped_value
```

```python
s = Stack(3)
```

```python
s.stack
```

```python
s.push(1)
s.push(2)
s.push(3)
```

```python
s.stack
```

```python
s.pop()
```
