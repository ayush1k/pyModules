# implementing stacks via Linked Lists

```python
class Node:
  def __init__(self, value):
    self.data = value
    self.next = None
```

```python
class Stack:

  def __init__(self):
    self.top = None
    self.size = 0

  def isEmpty(self):
    return self.top == None

  def push(self, value):
    new_node = Node(value)
    new_node.next = self.top
    self.top = new_node
    self.size += 1

  def traverse(self):
    curr = self.top
    while curr != None:
      print(curr.data)
      curr = curr.next

  def peek(self): # returns the value at top
    if (self.isEmpty()):
      return 'Stack is empty'
    return self.top.data

  def pop(self):
    if (self.isEmpty()):
      return 'Stack is empty'
    else:
      popped_value = self.top.data
      self.top = self.top.next
      self.size -= 1
      return popped_value
```

```python
# string reversal using stacks

def reverse_string(text):
  S =  Stack()
  for i in text:
    S.push(i)

  res = ''
  while (not S.isEmpty()):
    res += S.pop()

  return res
```

```python
def text_editor(text, pattern):
    u = Stack()
    r = Stack()

    for i in text:
        u.push(i)

    for i in pattern:

        if i == 'u':
            data = u.pop()
            r.push(data)
        else:
            data = r.pop()
            u.push(data)
    res = ''
    while (not u.isEmpty()):
       res = u.pop() + res

    return res
```
