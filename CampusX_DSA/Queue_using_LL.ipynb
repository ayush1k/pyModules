```python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None
```

```python
class Queue:

  def __init__(self):
    self.front = None
    self.rear = None

  def is_empty(self):
    return self.front == None

  def size(self):
    curr = self.front
    counter = 0
    while curr != None:
      counter += 1
      curr = curr.next
    return counter

  def enqueue(self, value):
    new_node = Node(value)
    if self.rear == None:
      self.front = new_node
      self.rear = self.front
    else:
      self.rear.next = new_node
      self.rear = new_node

  def dequeue(self):
    if self.front == None:
      return 'Empty'
    else:
      self.front = self.front.next

  def front_item(self):
    if self.front == None:
      return 'Empty'
    else:
      return self.front.value

  def rear_item(self):
    if self.front == None:
      return 'Empty'
    else:
      return self.rear.value

  def traverse(self):
    curr = self.front
    while curr != None:
      print(curr.value)
      curr = curr.next
```

```python
q = Queue()
```

```python
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
```

```python
q.traverse()
```
