### What is Encapsulation?

**Encapsulation** is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, or class. It also involves restricting direct access to some of an object's components, which means that the internal representation of an object is hidden from the outside. Instead, access is provided through defined methods.

Key aspects of encapsulation:

1.  **Bundling Data and Methods:** It groups related data and the functions that manipulate that data together within a class.
2.  **Data Hiding (Information Hiding):** It restricts direct access to an object's internal state. This is often achieved by making attributes *private* or *protected* (though Python's interpretation of 'private' is more of a convention).
3.  **Controlled Access:** External code interacts with the object's data only through publicly exposed methods, which provides control over how the data is accessed and modified. This helps maintain data integrity and prevents unintended modification.

In Python classes, the double underscore __ prefix has a special meaning and is used in two primary ways:

Name Mangling for 'Private' Attributes: When you prefix an attribute or method name with a double underscore (e.g., __attribute_name or __method_name) within a class, Python automatically 'mangles' its name. This means the interpreter internally changes the name to _ClassName__attribute_name.

Purpose: This makes it harder for subclasses to accidentally override these attributes/methods and makes it harder for external code to directly access them, effectively serving as a soft form of encapsulation or 'privacy'. It's a convention to signal that these members are intended for internal use within the class and should not be directly accessed or modified from outside.
Special Methods (Dunder Methods): Methods that begin and end with double underscores (e.g., __init__, __str__, __add__) are known as 'dunder' methods or 'magic' methods. These methods have special meaning to the Python interpreter and are typically called implicitly by Python in response to certain operations or events.

Purpose: They allow you to define how objects of your class behave with built-in operations, functions, and language constructs (e.g., how they are initialized, how they are converted to strings, how they respond to arithmetic operators like +, -, etc.).

### Tabulated Differences: Pass by Value, Pass by Reference, Python's Pass by Object Reference, and Call by Name

Let's summarize the key differences between these argument passing mechanisms:

| Feature                  | Pass by Value                                  | Pass by Reference                                | Python's Pass by Object Reference                | Call by Name                                    |
| :----------------------- | :--------------------------------------------- | :----------------------------------------------- | :----------------------------------------------- | :---------------------------------------------- |
| **Mechanism**            | A copy of the argument's value is passed.        | A reference (memory address) to the original variable is passed. | A copy of the reference to the object is passed. | The argument expression is textually substituted and re-evaluated at each use. |
| **Effect on Original**   | Original variable **unaffected** by function modifications. | Original variable **can be modified** by function. | Modifications to *mutable* objects affect original. Reassignment of parameter to new object for *immutable* types does **not** affect original. | Original variables **can be affected** if argument expression has side effects and is re-evaluated. |
| **Performance**          | Can be slower for large data structures due to copying. | Generally faster as only an address is passed.   | Efficient as only a reference copy is made.      | Potentially slower if expression is complex and evaluated many times. |
| **Readability/Safety**   | High safety, clear boundaries.                 | Less safe, function can alter caller's state.    | Good balance, depends on object mutability.      | Can be hard to reason about due to repeated evaluation and potential side effects. |
| **Common Languages**     | C, C++, Java (for primitives)                  | C++ (with `&`), Pascal, some Fortran           | Python, Java (for objects), Ruby                 | Algol 60, Scheme (with macros/special forms)    |

### Static Variables (Class Variables)

**Definition:** Static variables, also known as class variables, are variables that are shared by all instances of a class. They are defined directly within the class body, outside of any methods. They are associated with the class itself, rather than with any specific object (instance) of the class.

**Characteristics:**
- **Shared:** All objects of the class share the same copy of the static variable.
- **Accessed via Class or Instance:** They can be accessed using either the class name (`ClassName.variable_name`) or an instance of the class (`instance_name.variable_name`). However, it's generally good practice to access them via the class name when modifying them to make it clear you're changing the class-level attribute.
- **Use Cases:** Often used for constants, counters, or any data that is common to all objects of a particular class.

### Static Methods

**Definition:** Static methods are methods that belong to the class but do not operate on the instance of the class (like regular methods) or the class itself (like class methods). They don't have access to `self` (instance) or `cls` (class) as their first argument.

**Characteristics:**
- **No `self` or `cls`:** They don't take `self` or `cls` as their first parameter.
- **Decorated with `@staticmethod`:** You define them using the `@staticmethod` decorator.
- **Independent:** They are essentially regular functions that happen to be defined inside a class, usually because they logically belong to the class's scope or functionality, but don't need any class-specific or instance-specific data.
- **Use Cases:** Utility functions that don't need to interact with instance data or class data, but are logically grouped with the class.

### Class Methods (for comparison)

**Definition:** Class methods are methods that are bound to the class and receive the class itself as their first argument, conventionally named `cls`.

**Characteristics:**
- **`cls` as first argument:** They take `cls` (the class object) as their first parameter.
- **Decorated with `@classmethod`:** You define them using the `@classmethod` decorator.
- **Access Class State:** They can access and modify class-level attributes (static variables) and can be used to create alternative constructors.
- **Use Cases:** Factory methods to create instances of the class in different ways, or methods that operate on class-level data.

```python
class Pizza:
    MAX_CHEESE = 500 # Class variable

    def __init__(self, ingredients):
        self.ingredients = ingredients

    def __repr__(self):
        return f'Pizza({self.ingredients!r})'

    @classmethod
    def margherita(cls):
        return cls(['mozzarella', 'tomatoes']) # Uses cls to create an instance

    @classmethod
    def prosciutto(cls):
        return cls(['mozzarella', 'tomatoes', 'ham'])

# Create pizzas using class methods (alternative constructors)
pizza1 = Pizza.margherita()
pizza2 = Pizza.prosciutto()

print(pizza1)
print(pizza2)

# Accessing class variable via class method context (though not demonstrated here directly)
# Example: A class method could check Pizza.MAX_CHEESE
```

In object-oriented programming, especially in the context of classes, understanding the relationships between them is crucial for designing robust and maintainable software. Here are the primary types of relationships:

### 1. Inheritance (Is-A Relationship):

*   **Definition:** This is when one class (the subclass or derived class) acquires the properties and behaviors of another class (the superclass or base class). It represents a hierarchical relationship where the subclass is a specialized version of the superclass.
*   **Keyword in Python:** `class SubClass(SuperClass):`
*   **Example:** A `Car` *is a* `Vehicle`. A `Dog` *is an* `Animal`.
*   **Characteristics:**
    *   Subclasses inherit attributes and methods from their superclass.
    *   Subclasses can extend or override the inherited behavior.
    *   Promotes code reusability and establishes a natural hierarchy.

### 2. Association (Has-A Relationship - General):

*   **Definition:** A broad relationship where one class is somehow connected to another. It means that objects of one class use or interact with objects of another class. It's a very general term.
*   **Example:** A `Student` *has a* `Course` (they are related, but one doesn't 'own' the other in a strong sense). A `Person` *has an* `Address`.
*   **Characteristics:**
    *   Can be one-to-one, one-to-many, many-to-one, or many-to-many.
    *   The objects can exist independently.
    *   Often implemented by one object having another object as an attribute.

### 3. Aggregation (Has-A Relationship - Weak):

*   **Definition:** A specialized form of association that represents a "has-a" relationship where one object (the "whole" or aggregate) contains another object (the "part"). The key here is that the "part" can exist independently of the "whole". If the "whole" is destroyed, the "part" can still exist.
*   **Example:** A `Department` *has a* `Professor`. If the department closes, the professor still exists.
*   **Characteristics:**
    *   "Whole" and "part" have independent lifecycles.
    *   Represented by a hollow diamond on the "whole" side in UML.

### 4. Composition (Has-A Relationship - Strong):

*   **Definition:** A more specialized and stronger form of association than aggregation. It also represents a "has-a" relationship, but implies that the "part" cannot exist without the "whole". If the "whole" is destroyed, the "part" is also destroyed.
*   **Example:** A `Car` *has an* `Engine`. If the car is destroyed, its specific engine is also considered destroyed with it (it might be salvaged, but as a *part of that car*, it's gone).
*   **Characteristics:**
    *   "Part" depends entirely on the "whole" for its existence; they have the same lifecycle.
    *   Represented by a solid diamond on the "whole" side in UML.

### 5. Dependency (Uses-A Relationship):

*   **Definition:** A weaker relationship where one class uses another class. It means that a change in one class might affect another, but the relationship is not as structural as association or inheritance. It's often temporary or parameter-based.
*   **Example:** A `Function` *uses a* `UtilityClass` to perform a specific task. The function might take an object of `UtilityClass` as a parameter.
*   **Characteristics:**
    *   Often represented by dashed arrows in UML.
    *   Indicates that one class needs another to fulfill its responsibilities but doesn't store a direct reference to it as an attribute.

Understanding these relationships helps you model real-world problems more accurately and design systems that are modular, flexible, and easier to maintain.

### Inheritance: A Detailed Explanation

**Definition:**
Inheritance is a fundamental principle of object-oriented programming (OOP) that allows a new class (called the **subclass**, **derived class**, or **child class**) to inherit properties and behaviors (attributes and methods) from an existing class (called the **superclass**, **base class**, or **parent class**). It establishes an "Is-A" relationship between classes, meaning the subclass *is a* specialized type of the superclass.

**Core Concepts:**

1.  **Reusability:** The primary benefit of inheritance is code reusability. Instead of writing the same attributes and methods in multiple classes, they can be defined once in a superclass and then inherited by any number of subclasses.
2.  **Extension:** Subclasses can extend the functionality of their superclass by adding new attributes and methods that are specific to their specialized nature.
3.  **Specialization:** Inheritance allows for the creation of a hierarchy of classes, moving from more general concepts (superclass) to more specific ones (subclasses). Each subclass represents a more specialized version of its parent.
4.  **Overriding:** Subclasses can provide their own specific implementations for methods that are already defined in their superclass. This concept is known as method overriding. When an overridden method is called on an object of the subclass, the subclass's version is executed instead of the superclass's version.
5.  **Polymorphism:** Inheritance is a cornerstone of polymorphism, which means "many forms." Through inheritance, an object of a subclass can be treated as an object of its superclass. This allows for writing more generic and flexible code, as a single function or method can operate on objects of different, but related, types.

**Key Characteristics:**

*   **Hierarchical Structure:** Inheritance naturally leads to a tree-like or hierarchical structure of classes, where commonalities are factored into higher-level classes.
*   **Code Organization:** It helps in organizing code logically by grouping related functionalities.
*   **Maintainability:** Changes made in a superclass can automatically propagate to all its subclasses, simplifying maintenance. However, changes must be carefully considered to avoid breaking existing subclass functionalities.
*   **Reduced Redundancy:** By inheriting common features, it minimizes redundant code, making the codebase smaller and less prone to errors.
*   **Abstraction:** Superclasses can define abstract methods (methods declared but not implemented) that force subclasses to provide their own implementations, thereby enforcing a contract for behavior.

**Types of Inheritance (common categorizations):**

1.  **Single Inheritance:** A subclass inherits from only one superclass.
2.  **Multiple Inheritance:** A subclass inherits from multiple superclasses, combining their attributes and methods.
3.  **Multilevel Inheritance:** A subclass inherits from a superclass, which in turn inherits from another superclass, forming a chain.
4.  **Hierarchical Inheritance:** Multiple subclasses inherit from a single superclass.
5.  **Hybrid Inheritance:** A combination of two or more of the above types.

**Limitations and Considerations:**

*   **Tight Coupling:** Inheritance creates a strong coupling between the superclass and subclass. Changes in the superclass can sometimes unintentionally affect subclasses.
*   **Complexity (especially with Multiple Inheritance):** Multiple inheritance can lead to complex class hierarchies and potential ambiguity issues (like the "Diamond Problem" in some languages, though Python has a Method Resolution Order (MRO) to handle this).
*   **Design Challenges:** Deciding what should be a superclass and what should be a subclass, and when to use inheritance versus composition (where one class *has* an object of another class as an attribute), is a significant design challenge.

### Polymorphism: A Detailed Explanation

**Definition:**
Polymorphism, derived from Greek words "poly" (many) and "morph" (form), is a core concept in object-oriented programming (OOP) that allows objects to take on "many forms." In essence, it enables a single interface to represent different underlying forms (data types or classes). This means you can design a generic interface that can be used to handle a variety of data types, reducing complexity and increasing flexibility.

**Core Concepts:**

1.  **Single Interface, Multiple Implementations:** The fundamental idea is to have a consistent way of interacting with different objects that might behave differently. For example, a common operation like `draw()` could be applied to various shapes (circle, square, triangle), but each shape would have its own specific way of drawing itself.
2.  **"Many Forms" of Objects:** An object can be treated as an instance of its own class, an instance of its parent class, or an instance of any interface it implements. This flexibility allows objects of different classes to be used interchangeably through a common interface, provided they share that interface.
3.  **Dynamic Binding (Late Binding):** Polymorphism often relies on dynamic binding, where the specific method implementation to be executed is determined at runtime, not at compile time. This means that when a method is called on an object through a polymorphic reference, the correct version of the method (belonging to the actual type of the object) is invoked.

**Key Characteristics & Types:**

Polymorphism manifests in several forms, often categorized as:

1.  **Method Overriding (Runtime Polymorphism / Dynamic Polymorphism):**
    *   **Description:** This is the most common form of polymorphism in inheritance hierarchies. A subclass provides its own specific implementation for a method that is already defined in its superclass. When the method is called on an object, the version belonging to the object's actual class (even if accessed via a superclass reference) is executed.
    *   **"Is-A" Relationship:** It works hand-in-hand with inheritance, where a child object can be substituted for a parent object.

2.  **Method Overloading (Compile-Time Polymorphism / Static Polymorphism):**
    *   **Description:** This involves defining multiple methods within the *same* class (or across classes) with the *same name* but different parameters (number, type, or order of arguments). The compiler (or interpreter) determines which method to call based on the arguments provided at the time of the call.
    *   **Note:** While some languages (like Java, C++) support traditional method overloading, Python typically does not. In Python, if you define multiple methods with the same name, the last one defined will overwrite the previous ones. Python achieves similar flexibility often through default arguments, variable-length arguments (`*args`, `**kwargs`), or type checking within a single method.

3.  **Duck Typing (Ad-hoc Polymorphism - common in Python):**
    *   **Description:** A concept particularly prominent in dynamically-typed languages like Python. It states: "If it walks like a duck and quacks like a duck, then it must be a duck." This means that the type or class of an object is less important than *what methods it has* (its behavior). If an object has the necessary methods, it can be used in a particular context, regardless of its explicit class hierarchy.
    *   **Focus on Behavior:** It emphasizes interfaces over strict class types, allowing for flexible and less coupled code.

**Benefits of Polymorphism:**

*   **Flexibility and Extensibility:** New classes can be added to the system without requiring changes to existing code, as long as they adhere to the common interface. This makes systems easier to extend and maintain.
*   **Code Reusability:** Generic code can be written that works with various types of objects, as long as they provide the required methods.
*   **Simplified Code:** Reduces the need for complex `if-elif-else` structures or `switch` statements to handle different object types, leading to cleaner and more readable code.
*   **Decoupling:** Decreases the dependency between different parts of the code. A client interacting with an object through a polymorphic interface doesn't need to know the specific type of the object, only that it supports the expected operations.
*   **Abstraction:** Helps in achieving higher levels of abstraction by focusing on the "what" (the behavior) rather than the "how" (the specific implementation).

In summary, polymorphism empowers developers to write more adaptable, reusable, and manageable code by allowing objects of different classes to be treated through a common interface.

### The `super()` Keyword: A Detailed Explanation

In Python, the `super()` keyword serves a crucial role in managing inheritance, particularly when dealing with method overriding and constructor chaining. It provides a way to access methods and attributes of a parent or sibling class from within a child class, enabling proper cooperation in complex class hierarchies.

**Primary Purpose:**

The fundamental purpose of `super()` is to give you access to the inherited methods that have been overridden in a subclass. It allows a subclass to call an implementation of a method or an attribute that resides higher up in the Method Resolution Order (MRO) of the class hierarchy.

**How it Works:**

When `super()` is called (typically without arguments, or with `super(ClassName, self)` in older Python 2 syntax or in specific advanced scenarios), it returns a **proxy object**. This proxy object then delegates method calls to the appropriate parent or sibling class according to the class's MRO. It's important to understand that `super()` does *not* return the parent class itself, nor does it necessarily return the *immediate* parent in the inheritance tree. Instead, it consults the MRO to find the next class in the inheritance chain that implements the method being called.

**Common Use Cases:**

1.  **Calling Parent Constructor (`__init__`)**: This is the most frequent use. When a child class defines its own `__init__` method, it often needs to ensure that the parent class's `__init__` method is also executed to properly initialize inherited attributes. Using `super().__init__(...)` ensures that the parent's initialization logic runs.
2.  **Calling Overridden Methods**: If a child class overrides a method from its parent, but also wants to incorporate the parent's functionality in addition to its own, it can call `super().method_name(...)` to invoke the parent's version of the method.
3.  **Cooperative Multiple Inheritance**: In scenarios with multiple inheritance (where a class inherits from more than one parent), `super()` becomes indispensable. It ensures that methods are called correctly across the inheritance graph in a predictable order, following the MRO. This allows for a cooperative approach where each class in the hierarchy can contribute to a method's implementation without needing to explicitly know about all other classes in the chain.

**Key Characteristics & Benefits:**

*   **Clearer Code:** It makes inheritance clearer and more maintainable by avoiding hardcoding parent class names. If the inheritance hierarchy changes, `super()` automatically adjusts.
*   **Method Resolution Order (MRO):** `super()` strictly adheres to the MRO, which is a linearization of a class's inheritance hierarchy. This predictable order prevents ambiguity and ensures that methods are called correctly, especially in complex multiple inheritance scenarios.
*   **Cooperative Behavior:** It facilitates cooperative multiple inheritance, allowing different base classes to contribute to the functionality of a method without causing conflicts.
*   **Consistency:** Promotes a consistent way of interacting with inherited classes.

In essence, `super()` acts as a powerful tool for managing method dispatch in class hierarchies, ensuring that all relevant base class implementations are considered when a method is called, leading to more robust and flexible object-oriented designs.

### Types of Inheritance: A Detailed Explanation

Inheritance, a cornerstone of object-oriented programming, allows classes to inherit properties and behaviors from other classes, forming a hierarchical structure. There are several distinct types of inheritance, each serving different design purposes:

#### 1. Single Inheritance

*   **Definition:** This is the simplest form of inheritance, where a subclass inherits from only one direct superclass. It creates a straightforward parent-child relationship.
*   **Structure:** A -> B (where B inherits from A).
*   **Characteristics:**
    *   Promotes clear and unambiguous class hierarchies.
    *   Easy to understand and implement.
    *   Reduces complexity compared to other forms of inheritance.
    *   Very common in many object-oriented languages.

#### 2. Multiple Inheritance

*   **Definition:** In multiple inheritance, a subclass can inherit from two or more direct superclasses. This allows the child class to combine features and behaviors from multiple independent sources.
*   **Structure:** A, B -> C (where C inherits from both A and B).
*   **Characteristics:**
    *   Can be very powerful for integrating diverse functionalities into a single class.
    *   Can lead to complex class hierarchies and potential ambiguities, often referred to as the "Diamond Problem" (where a class inherits from two classes that both inherit from a common base class, leading to uncertainty about which inherited method to use).
    *   Some languages (like Java) do not directly support multiple inheritance for classes (though they do for interfaces), while others (like Python, C++) do, often with mechanisms to resolve ambiguities (e.g., Method Resolution Order in Python).

#### 3. Multilevel Inheritance

*   **Definition:** Multilevel inheritance involves a chain of inheritance, where a class inherits from a base class, and then another class inherits from that derived class. It forms a hierarchical structure where one class acts as an intermediate base class.
*   **Structure:** A -> B -> C (where C inherits from B, and B inherits from A).
*   **Characteristics:**
    *   Extends the concept of single inheritance over multiple levels.
    *   Allows for deeper specialization and abstraction.
    *   Each level in the hierarchy can add its own specific features or modify inherited ones.
    *   Can become complex if the chain becomes too long, potentially leading to tight coupling across many levels.

#### 4. Hierarchical Inheritance

*   **Definition:** In hierarchical inheritance, a single superclass is inherited by multiple subclasses. This means one parent class has several child classes, each specializing in a different way.
*   **Structure:** B, C -> A (where B inherits from A, and C also inherits from A).
*   **Characteristics:**
    *   Useful when a common base concept needs to be specialized into several distinct variations.
    *   Promotes code reuse for common functionalities defined in the base class.
    *   Each subclass develops its own unique characteristics while sharing a common foundation.

#### 5. Hybrid Inheritance

*   **Definition:** Hybrid inheritance is a combination of two or more of the above types of inheritance within a single class hierarchy. It allows for highly flexible and complex class designs.
*   **Structure:** A combination of any of the above patterns (e.g., a class might use multiple inheritance, and one of its parents might be part of a multilevel inheritance chain).
*   **Characteristics:**
    *   Offers maximum flexibility in designing class relationships.
    *   Can be extremely powerful for modeling intricate real-world systems.
    *   Due to its combined nature, it can also introduce significant complexity and potential design challenges, including ambiguity issues that need careful management (especially in languages that support multiple inheritance).

Understanding these different types is crucial for choosing the most appropriate design pattern for a given problem, balancing the benefits of code reuse and extensibility with the need for clarity, maintainability, and avoiding unnecessary complexity.

### Method Overloading: A Detailed Explanation

**Definition:**
Method overloading refers to the ability to define multiple methods within the *same class* that share the *same name* but have different parameters. The distinction between these methods is typically based on the number, type, or order of their arguments.

**Traditional View (in languages like Java, C++):**
In languages that strictly support method overloading, the compiler or interpreter can differentiate between methods with the same name by analyzing the method signature (the method name plus the types and number of its parameters). When a method is called, the system automatically picks the correct version based on the arguments provided.

**Python's Approach:**
Python does **not** support method overloading in the traditional sense. If you define multiple methods with the same name within a single class, the *last one defined will effectively overwrite all previous definitions* of that method. When you call the method, only the most recently defined version will be executed.

However, Python provides alternative mechanisms to achieve similar flexibility and behavior often associated with method overloading:

1.  **Default Arguments:** You can make parameters optional by assigning default values. This allows a single method definition to handle calls with varying numbers of arguments.
2.  **Variable-Length Arguments (`*args` and `**kwargs`):**
    *   `*args` (arbitrary positional arguments) allows a method to accept any number of non-keyword arguments.
    *   `**kwargs` (arbitrary keyword arguments) allows a method to accept any number of keyword arguments.
    Using these, a single method can be designed to handle diverse argument lists dynamically.
3.  **Type Checking (within the method):** You can implement conditional logic (e.g., `if-elif-else` statements) inside a single method to check the type or number of arguments passed and perform different actions accordingly.

**Conclusion for Python:** While Python lacks direct syntax for method overloading, its dynamic nature and flexible argument handling mechanisms allow developers to write highly adaptable methods that can respond differently based on the input, effectively serving the same purpose as method overloading in other languages.


### Operator Overloading: A Detailed Explanation

**Definition:**
Operator overloading is a feature that allows you to redefine the behavior of standard operators (like `+`, `-`, `*`, `/`, `==`, `>`, `[]`, etc.) when they are used with instances of custom classes. This means you can specify what an operator should do when applied to objects of your own design.

**Purpose:**

*   **Intuitive Code:** It makes code more readable and intuitive by allowing custom objects to interact with operators in a way that is consistent with their real-world or logical meaning. For example, it makes sense to "add" two `Fraction` objects using the `+` operator.
*   **Consistency:** It enables custom objects to behave more like built-in data types, integrating seamlessly into expressions and algorithms.

**How it Works (Magic Methods / Dunder Methods):**
In Python, operator overloading is achieved by implementing special methods, often referred to as "magic methods" or "dunder methods" (due to their double underscores, e.g., `__add__`, `__sub__`, `__mul__`). Each standard operator corresponds to a specific magic method within a class.

When an operator is used with objects of a custom class, Python automatically calls the corresponding magic method defined in that class. For example:

*   When you use the `+` operator (e.g., `obj1 + obj2`), Python calls `obj1.__add__(obj2)`.
*   When you use `==` (e.g., `obj1 == obj2`), Python calls `obj1.__eq__(obj2)`.
*   When you use `len()` (e.g., `len(obj)`), Python calls `obj.__len__()`.

By providing an implementation for these magic methods within your class, you dictate how your objects will respond to those operators.

**Examples of Operator Overloading Capabilities:**

*   **Arithmetic Operators:** `__add__` (`+`), `__sub__` (`-`), `__mul__` (`*`), `__truediv__` (`/`), `__floordiv__` (`//`), `__mod__` (`%`), `__pow__` (`**`).
*   **Comparison Operators:** `__eq__` (`==`), `__ne__` (`!=`), `__lt__` (`<`), `__le__` (`<=`), `__gt__` (`>`), `__ge__` (`>=`).
*   **Unary Operators:** `__pos__` (`+obj`), `__neg__` (`-obj`), `__abs__` (`abs(obj)`).
*   **Container Operators:** `__len__` (`len(obj)`), `__getitem__` (`obj[key]`), `__setitem__` (`obj[key] = value`), `__delitem__` (`del obj[key]`), `__contains__` (`item in obj`).
*   **Type Conversion:** `__str__` (for `str(obj)` or `print(obj)`), `__repr__` (for `repr(obj)`).

**Benefits:**

*   **Readability:** Code becomes more natural and easier to understand when custom objects can use standard operators.
*   **Code Reusability:** It allows you to reuse existing algorithms and patterns that rely on operators, even with your custom types.
*   **Abstraction:** It hides the internal implementation details of how operations are performed on your objects, presenting a clean interface.

Operator overloading is a powerful tool in Python that significantly enhances the expressiveness and usability of custom classes, making them behave more consistently with built-in types.
